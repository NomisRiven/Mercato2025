```python
df_TOP = df_TOP.drop(columns=["Country"])
df_TOP.head()
```

-------------------

```python
Nb_games = 10
KDA = 2
DPM = 500
CSD_15 = 5
```

```python
df_TOP_temp = df_TOP.copy()
df_TOP_temp["Games"] = pd.to_numeric(df_TOP_temp["Games"], errors='coerce')
df_TOP_temp["KDA"] = pd.to_numeric(df_TOP_temp["KDA"], errors='coerce')
df_TOP_temp["DPM"] = pd.to_numeric(df_TOP_temp["DPM"], errors='coerce')
df_TOP_temp["CSD@15"] = pd.to_numeric(df_TOP_temp["CSD@15"], errors='coerce')

df_TOP_sorted = df_TOP_temp.sort_values(by=["Games", "KDA", "DPM", "CSD@15"], ascending=False)

df_TOP_filtered = df_TOP_sorted[
    (df_TOP_sorted["Games"] >= Nb_games) &
    (df_TOP_sorted["KDA"] >= KDA) &
    (df_TOP_sorted["DPM"] >= DPM) &
    (df_TOP_sorted["CSD@15"] >= CSD_15)
]

print("\nJoueurs répondant à tous les critères :")
df_TOP_filtered
```
